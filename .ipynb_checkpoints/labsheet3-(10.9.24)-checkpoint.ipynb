# create embedding and find pairwise similarity 

```python
import numpy as np
import os
```

```python
from openai import OpenAI
```

```python
client=OpenAI()
```

```python
texts=["I love reading","I am angry","I play tennsi","I am tall"]
```

```python
response=client.embeddings.create( model="text-embedding-ada-002",input=texts)
```

```python
for i in range(len(texts)-1):
    embd1=response.data[i].embedding
    for j in range(i+1,len(texts)):
        embd2=response.data[i].embedding
        sim_score=np.dot(embd1,embd2)*100
        print("Similarity between text {0} and text {1} is {2}%".format(texts[i],texts[j],sim_score) )
```

# GIVEN A SET OF DOCUMENTS AND A PLAGIARISED TEXT, FIND 

```python
import numpy as np
from openai import OpenAI
client=OpenAI()
```

```python
doc1=" Renewable energy refers to the energy harnessed from natural resources that are constantly replenished, such as solar, wind, hydro, and geothermal. These energy sources are considered sustainable alternatives to fossil fuels, offering reduced greenhouse gas emissions and minimized environmental impacts. Advanced technologies, including photovoltaic systems and wind turbines, have enhanced efficiency and grid integration. The future of renewable energy hinges on innovations like energy storage systems, smart grids, and decentralized generation, transforming the global energy landscape."
```

```python
doc2="Artificial Intelligence (AI) is a branch of computer science that aims to create machines capable of performing tasks that typically require human intelligence, such as learning, reasoning, and problem-solving. Recent advances in machine learning, neural networks, and natural language processing have enabled AI systems to excel in areas like autonomous driving, healthcare diagnostics, and personalized recommendations. As AI evolves, ethical concerns related to privacy, bias, and job displacement emerge, necessitating responsible innovation and regulation to balance progress and societal impacts. "
```

```python
doc3= "Marine biodiversity encompasses the variety of life forms in oceanic and coastal ecosystems, including species, genetic variation, and ecosystems such as coral reefs, mangroves, and the deep sea. Marine organisms play critical roles in nutrient cycling, climate regulation, and food chains, with many species contributing to ecosystem resilience. However, anthropogenic threats like overfishing, pollution, and climate change have led to significant biodiversity loss. Preserving marine biodiversity requires global efforts in marine conservation, habitat restoration, and sustainable fisheries management. " 
```

```python
plag_text="Renewable energy is derived from resources that are replenished naturally, such as sunlight, wind, and water. These sources offer a sustainable alternative to fossil fuels, contributing to the reduction of carbon emissions and environmental damage. Photovoltaic technology and wind turbines have increased efficiency, supporting the transition to a low-carbon energy system. Hydropower remains a key player, providing large-scale energy from river and oceanic currents. Geothermal energy taps into the Earth’s internal heat for a consistent power source. As storage technologies improve, such as battery systems and hydrogen fuel cells, renewable energy will be able to supply more stable power, reducing reliance on non-renewable resources. Governments worldwide are pushing policies and incentives to accelerate the adoption of renewable energy, recognizing its role in mitigating climate change and fostering sustainable development. "
```

```python
docs=[plag_text,doc1,doc2,doc3]
```

```python
response=client.embeddings.create(
    model= "text-embedding-ada-002",
    input=docs
)
plag_embd=response.data[0].embedding
```

```python
for i in range(1,len(docs)):
    embd=response.data[i].embedding
    sim_score=np.dot(plag_embd,embd)*100
    print("similarity between plag_text and doc{0} is {1}%".format(i,sim_score))
```
